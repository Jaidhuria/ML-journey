```python
#heat map
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np

# Load dataset
var = sns.load_dataset("penguins")
# w
# Drop rows with missing values for accurate correlation
var_clean = var.dropna()

# Compute the correlation matrix
corr = var_clean.corr(numeric_only=True)

# Plot the heatmap
sns.heatmap(corr, annot=True, cmap="Purples")
plt.title("Correlation Heatmap - Penguins Dataset")
plt.show()
```
