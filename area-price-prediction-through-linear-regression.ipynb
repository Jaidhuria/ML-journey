```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import linear_model
import seaborn as sns

# Load data
df = pd.read_csv("/kaggle/input/homeprices/homeprices.csv")

# Train the model
new_df = df.drop('price', axis="columns")
model = linear_model.LinearRegression()
model.fit(new_df, df.price)

# Make a prediction
area_to_predict = 1500
predicted_price = model.predict([[area_to_predict]])[0]

# Plot existing data
plt.figure(figsize=(10,6))
sns.lineplot(x="area", y="price", data=df, palette="Blues")

# Add predicted value as a red dot
plt.scatter(x=[df['area'].max()], y=[predicted_price], color='black', s=100, label=f'Predicted price for {area_to_predict} sqft')
plt.legend()
plt.title("House Price Prediction")
plt.xlabel("Area (sqft)")
plt.ylabel("Price")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import linear_model
import seaborn as sns

# Load data
df = pd.read_csv("/kaggle/input/homeprices/homeprices.csv")

# Train the model
new_df = df.drop('price', axis="columns")
model = linear_model.LinearRegression()
model.fit(new_df, df.price)

# Make a prediction
area_to_predict = 5500
predicted_price = model.predict([[area_to_predict]])[0]

# Plot existing data
plt.figure(figsize=(10,6))
sns.lineplot(x="area", y="price", data=df, marker='o')

# Add predicted value as a red dot
plt.scatter(x=[area_to_predict], y=[predicted_price], color='black', s=100, label='Predicted Price')

# Annotate the predicted value
plt.text(area_to_predict, predicted_price + 10000, f'{int(predicted_price):,}', color='skyblue', fontsize=12, ha='center')

# Labels and titles
plt.legend()
plt.title("House Price Prediction")
plt.xlabel("Area (sqft)")
plt.ylabel("Price")
plt.grid(True)
plt.show()
```
